```python
for label in df.columns:
  plt.hist(df[df["median_house_value"]>=50000][label],color='blue',label='one',alpha=0.7,density=True)
  plt.hist(df[df["median_house_value"]<=50000][label],color='red',label='one',alpha=0.7,density=True)
  plt.title(label)
  plt.xlabel(label)
  plt.ylabel('probability')
  plt.legend()
  plt.show()
```
